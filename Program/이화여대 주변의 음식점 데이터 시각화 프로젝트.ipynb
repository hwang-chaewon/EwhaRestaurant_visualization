#### **<i>[ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ëª…]</i>**

**1) folium library**  : ì§€ë„ë¥¼ í™œìš©í•œ ìœ„ì¹˜ ì‹œê°í™” ë¼ì´ë¸ŒëŸ¬ë¦¬  
**<i>-Marker</i>** :ë§ˆì»¤ë¥¼ ì„¤ì •í•´ì¤ë‹ˆë‹¤    
**<i>-MarkerCluster</i>** : ë§ˆì»¤ë“¤ì„ ë¬¶ìŒìœ¼ë¡œ ë³´ì—¬ì¤ë‹ˆë‹¤.


**2) geopy library**: ì¥ì†Œì˜ ì£¼ì†Œ, ìœ„ë„, ê²½ë„, ê³ ë„, ê±°ë¦¬ ê³„ì‚° ë“±ì˜ ê¸°ëŠ¥ì„ ì‰½ê²Œ ì´ìš©í•  ìˆ˜ ìˆê²Œ í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬  
**<i>-geodesic()</i>**: ì§€êµ¬ì—ì„œ ë‘ ì¥ì†Œ ì‚¬ì´ ê°€ì¥ ì§§ì€ ê±°ë¦¬ë¥¼ ê³„ì‚°í•´ì¤ë‹ˆë‹¤

```python
#!pip install folium
import folium
from folium.plugins import MarkerCluster
from geopy.distance import geodesic

def program_1():
  data_path = '/content/gdrive/MyDrive/á„‘á…¡á„ƒá…¦á„‡á…®á†«á„‘á…³á†¯á„Œá…¦á†¨/ìµœì¢…/ë°ì´í„°/'
  df = pd.read_csv(data_path + 'final_all.csv', encoding = "utf-8")
  dong = sorted(set(list((df['í–‰ì •ë™ëª…']))))
  print('ì„œëŒ€ë¬¸êµ¬ì˜ í–‰ì •ë™ëª…ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤ : \n', dong)
  print()
  region = input('ì›í•˜ëŠ” í–‰ì •ë™ëª…ì„ ì…ë ¥í•˜ì„¸ìš”: \n')
  print()

  while True:
      df = df.loc[df['í–‰ì •ë™ëª…'] == region]
      mode_1 = int(input('ëª¨ë“œë¥¼ ê³¨ë¼ì£¼ì„¸ìš”: 1 - í•´ë‹¹ ë™ ì „ì²´ ë·° ëª¨ë“œ, 2 - ê²€ìƒ‰ ëª¨ë“œ: , 3 - í”„ë¡œê·¸ë¨ ì¢…ë£Œ: \n'))
      if mode_1 == 1:
        
          lat = df['ìœ„ë„'].mean()
          long = df['ê²½ë„'].mean()
          m1 = folium.Map(location = [lat, long], zoom_start = 13, width='70%', height='70%', position='relative')
          marker_cluster=MarkerCluster().add_to(m1)
          for i in df.index:
            sub_lat = df.loc[i,'ìœ„ë„']
            sub_long = df.loc[i,'ê²½ë„']

        #Markerê¸°ëŠ¥ìœ¼ë¡œ tooltip, icon, popupì„ ì§€ì •í•˜ì—¬ ì‹œê°í™” íš¨ê³¼ë¥¼ ë†’ì´ê³  ì£¼ì†Œ, ë‚˜ì™€ì˜ ê±°ë¦¬ë¥¼ ì œê³µí•©ë‹ˆë‹¤                
        #color: ë§ˆì»¤ì˜ ìƒ‰ê¹”  
        #icon: iconì¢…ë¥˜  
        #tooltip: ë§ˆìš°ìŠ¤ ì»¤ì„œë¥¼ ë§ˆì»¤ì— ì˜¬ë ¸ì„ ë•Œ ëœ¨ëŠ” ë¬¸êµ¬  
        #popup: ë§ˆìš°ìŠ¤ë¡œ ë§ˆì»¤ë¥¼ í´ë¦­í–ˆì„ë•Œ ëœ¨ëŠ” ë¬¸êµ¬ 
            folium.Marker([sub_lat, sub_long],
                          icon = folium.Icon(color='orange', icon = 'glyphicon-cutlery'),
                          tooltip = '<i><b>\'' + str(df.loc[i,'ìƒí˜¸ëª…']) + '\'</b></i>').add_to(marker_cluster)
          print('____________________________________________________')
          print('1ë²ˆ: í•´ë‹¹ ë™ ì „ì²´ ë·° ëª¨ë“œë¡œ ì„¤ì •ë˜ì—ˆìŠµë‹ˆë‹¤.')
          print()
          print(region + 'ì˜ ìŒì‹ì  ë¶„í¬ì…ë‹ˆë‹¤')
          display(m1)
          folium.Map().get_root().render()

            
      elif mode_1 == 2: #ê²€ìƒ‰ ëª¨ë“œì˜ ê²½ìš°, ë‚˜ì˜ ìœ„ì¹˜ì™€ ìŒì‹ì ì„ ì…ë ¥í•´ ì›í•˜ëŠ” ìŒì‹ì ê³¼ì˜ ê±°ë¦¬ë¥¼ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤

            print('____________________________________________________')
            print('2ë²ˆ: ê²€ìƒ‰ ëª¨ë“œë¡œ ì„¤ì •ë˜ì—ˆìŠµë‹ˆë‹¤')
            print()
            print('ìŒì‹ì ê³¼ í˜„ì¬ ìœ„ì¹˜ ì‚¬ì´ ê±°ë¦¬ë¥¼ ì•Œê¸° ìœ„í•´ ìœ„ì¹˜ ì…ë ¥ì´ í•„ìš”í•©ë‹ˆë‹¤')
            print('ë‚˜ì˜ ìœ„ì¹˜ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”: ')
            print()

            #ë‚˜ì˜ í˜„ì¬ ìœ„ì¹˜ë¥¼ ìœ„ë„ì™€ ê²½ë„ ì¢Œí‘œë¡œ ì…ë ¥í•©ë‹ˆë‹¤
            my_lat = float(input('ìœ„ë„ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”(-90~90): \n')) 
            my_long = float(input('ê²½ë„ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”(-180~180): \n')) 
            print()
            my_location = (my_lat, my_long)

            #ê²€ìƒ‰í•˜ê³ ì í•˜ëŠ” ìŒì‹ì  ì´ë¦„ì„ ì…ë ¥í•©ë‹ˆë‹¤
            store = input("ì£¼ì†Œì™€ ë‚˜ì™€ì˜ ê±°ë¦¬ë¥¼ ì•Œê³  ì‹¶ì€ ìŒì‹ì  ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”(í‚¤ì›Œë“œ ê²€ìƒ‰): \n")

            #ì…ë ¥í•œ ìŒì‹ì ì˜ ìœ„ë„, ê²½ë„ ì¢Œí‘œë¥¼ ì €ì¥í•©ë‹ˆë‹¤
            shop_list = df['ìƒí˜¸ëª…'].values.tolist() 
            stores = []
            for shop in shop_list:
              if store in shop:
                stores.append(shop)
            if len(stores) == 0:
              print("ì…ë ¥í•˜ì‹  í‚¤ì›Œë“œ {}(ì´)ê°€ ë“¤ì–´ê°„ ì‹ë‹¹ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.".format(store))
              continue
            dic = dict(zip(np.arange(len(stores)),stores))
            print("ê²€ìƒ‰í•˜ì‹  í‚¤ì›Œë“œì— ë§ëŠ” ì‹ë‹¹ ëª©ë¡ì…ë‹ˆë‹¤.")
            print(dic)
            num = int(input("ëª©ë¡ ì¤‘ ì›í•˜ì‹œëŠ” ì‹ë‹¹ ë²ˆí˜¸ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”: \n"))
            #print(dic.values)
            i = df[df['ìƒí˜¸ëª…'] == dic[num]].index.values
            i = int(i[0])

            store_location = (df['ìœ„ë„'][i],df['ê²½ë„'][i])
            
            #geodesicí•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ ë‚´ ìœ„ì¹˜ì™€, ì…ë ¥í•œ ë™ì— ìˆëŠ” ëª¨ë“  ìŒì‹ì  ì‚¬ì´ì˜ ê±°ë¦¬ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤
            df['distance']=df.apply(lambda x:geodesic((x['ìœ„ë„'],x['ê²½ë„']),my_location),axis=1) #axis=1ì„ í†µí•´ ì—´ ë°©í–¥ìœ¼ë¡œ ê±°ë¦¬ ê³„ì‚°ì„ ìˆ˜í–‰í•˜ê³ , ê·¸ ê²°ê³¼ë¥¼ dfì— 'distance' columnìœ¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤
            m2=folium.Map(location = [(store_location[0] + my_location[0]) / 2, (store_location[1] + my_location[1]) / 2], zoom_start=12,  position='relative')
      
    
            #Markerê¸°ëŠ¥ìœ¼ë¡œ tooltip, icon, popupì„ ì§€ì •í•˜ì—¬ ì‹œê°í™” íš¨ê³¼ë¥¼ ë†’ì´ê³  ì£¼ì†Œ, ë‚˜ì™€ì˜ ê±°ë¦¬ë¥¼ ì œê³µí•©ë‹ˆë‹¤
            folium.Marker(store_location, tooltip= '<b><i>'+str(df['ìƒí˜¸ëª…'][i])+'</b></i>',
                          icon=folium.Icon(color='blue', icon='search'),
                          popup='<i>\'ë„ë¡œëª…ì£¼ì†Œ\'</i>' + '<br/>' + df['ë„ë¡œëª…ì£¼ì†Œ'][i] +  '<br/>'*2 + 
                          '<i>\'ë³„ì \'</i>' + '<br/>' + str(df['naver_visitor_review_rate'][i])
                          + '<br/>'* 2  + '<i>\'ë‚˜ì™€ì˜ ê±°ë¦¬\'</i>' + '<br/>' + str(df['distance'][i])).add_to(m2)

            folium.Marker(my_location, tooltip = '<i><b>\'ë‚˜ì˜ ìœ„ì¹˜\'</b></i>', icon = folium.Icon(color = 'red', icon = 'user')).add_to(m2)
            print('<',dic[num],'>ì˜ ì£¼ì†Œ, ê±°ë¦¬ì— ëŒ€í•œ ì •ë³´ì…ë‹ˆë‹¤.')

            
            #ì‹œê°í™”ê°€ ì™„ì„±ëœ ì§€ë„ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤
            display(m2)
            folium.Map().get_root().render()
            
            
      elif mode_1 == 3: #í”„ë¡œê·¸ë¨ ì¢…ë£Œë¥¼ ì„ íƒí•  ê²½ìš° breakë¥¼ í†µí•´ ë°˜ë³µë¬¸ì„ ì¢…ë£Œí•©ë‹ˆë‹¤
            print()
            print('ì´ìš©í•´ ì£¼ì…”ì„œ ê°ì‚¬í•©ë‹ˆë‹¤')
            break
```

### **5. <i>'ë¦¬ë·° í…ìŠ¤íŠ¸'</i> íŒŒì´ ì°¨íŠ¸ ì‹œê°í™”**

í¬ë¡¤ë§í•œ ë¦¬ë·° ë°ì´í„°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•˜ì—¬ ì‚¬ìš©ìê°€ ì›í•˜ëŠ” ì§€ì—­ê³¼ í‚¤ì›Œë“œ ì„¸ ê°œë¥¼ ì…ë ¥í•˜ë©´,   
í‚¤ì›Œë“œ ì„¸ ê°œì˜ ë¹ˆë„ìˆ˜ê°€ ê°€ì¥ ë†’ì€ ì‹ë‹¹ì„ ì°¾ì•„ í•´ë‹¹ í‚¤ì›Œë“œ ê°ê°ì´ ì „ì²´ ë¦¬ë·°ì—ì„œ ëª‡ í¼ì„¼íŠ¸ë¥¼ ì°¨ì§€í•˜ê³  ìˆëŠ”ì§€ íŒŒì´ì°¨íŠ¸ë¡œ ì‹œê°í™”í•˜ì—¬ ë³´ì—¬ì¤€ë‹¤.

#### **<i>[ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ëª…]</i>**

**1)nltk**  : ìì—°ì–´ì²˜ë¦¬ ë¼ì´ë¸ŒëŸ¬ë¦¬, FreqDistë¥¼ í™œìš©í•˜ì—¬ í…ìŠ¤íŠ¸ì—ì„œ ë‹¨ì–´ì˜ ì¶œí˜„ íšŸìˆ˜ë¥¼ ì•Œë ¤ì¤€ë‹¤.   
**2)konlpy**  : ìì—°ì–´ì²˜ë¦¬ ë¼ì´ë¸ŒëŸ¬ë¦¬, í…ìŠ¤íŠ¸ë¥¼ ë„£ìœ¼ë©´ í˜•íƒœì†Œë³„ë¡œ ë‚˜ëˆ„ì–´ì„œ í’ˆì‚¬ë¥¼ ë¶„ì„í•´ì¤€ë‹¤.



```python
# !pip install konlpy
from konlpy.tag import Okt 
from nltk import FreqDist


# ë¦¬ë·°ì—ì„œ ëª…ì‚¬ ì¶”ì¶œí•´ì£¼ëŠ” í•¨ìˆ˜, ì§ì ‘ì ìœ¼ë¡œ ì‚¬ìš©í•˜ì§€ ì•Šê³  ë°‘ì˜ update_tokensì—ì„œ ë°›ì•„ì„œ ì‚¬ìš©
def tokenizer(review):
  okt = Okt()
  nouns = okt.nouns(review)      # Okt ì‚¬ìš©í•˜ì—¬ í† í°í™”í•œ í›„ ëª…ì‚¬ë§Œ ì¶”ì¶œ
  return [noun for noun in nouns]
# ì‚¬ìš©ìê°€ ì›í•˜ëŠ” ì§€ì—­ì˜ ì •ë³´ë¥¼ ê°€ì ¸ì™€ì„œ tokenize
def update_tokens():
  a=[]
  data_path = '/content/gdrive/MyDrive/á„‘á…¡á„ƒá…¦á„‡á…®á†«á„‘á…³á†¯á„Œá…¦á†¨/ìµœì¢…/ë°ì´í„°/'
  df = pd.read_csv(data_path + 'final_all.csv', encoding = "utf-8")
  dong = sorted(set(list((df['í–‰ì •ë™ëª…']))))
  print('ì„œëŒ€ë¬¸êµ¬ì˜ í–‰ì •ë™ëª…ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤ : \n', dong)
  region = input('ì›í•˜ëŠ” ì§€ì—­ì„ ì…ë ¥í•˜ì„¸ìš”:\n ')
  
  # ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
  df = df.loc[df['í–‰ì •ë™ëª…'] == region]
  columns = ['ìƒí˜¸ëª…', 'í–‰ì •ë™ëª…', 'naver_visitor_review_text', 'google_review']
  df = df[columns].copy()
  # ë¦¬ë·°ê°€ ë¹„ì—ˆìœ¼ë©´ ''ë¡œ ë°”ê¿ˆ
  df['naver_visitor_review_text'].replace(np.nan, '', inplace=True)
  df['google_review'].replace(np.nan, '', inplace=True)           
  
  # ë¦¬ë·°ë¥¼ tokenizer í•¨ìˆ˜ë¡œ í† í°í™”í•˜ì—¬ ë°ì´í„°í”„ë ˆì„ì— 'tokens' ì—´ë¡œ ì¶”ê°€
  for i in df.index:
    a.append(tokenizer(df['naver_visitor_review_text'][i]) + tokenizer(df['google_review'][i]))
  df['tokens'] = a
  
  # ë¹ˆë„ìˆ˜ê°€ ë†’ì€ ë‹¨ì–´ 5ê°œë¥¼ ë°ì´í„°í”„ë ˆì„ 'keys' ì—´ë¡œ ì¶”ê°€
  words = []
  for i in df.index:
    b = []
    fd = FreqDist(df['tokens'][i])
    commons = fd.most_common(5)       # ê° ì‹ë‹¹ì˜ tokens ì¤‘ ê°€ì¥ ë¹ˆë„ìˆ˜ê°€ ë†’ì€ 5ê°œë¥¼ ì°¾ì•„ì„œ ë‹¨ì–´ì™€ ë¹ˆë„ìˆ˜ íŠœí”Œë¡œ ì¶œë ¥. ex> (('ë§›ì§‘', 17), ('ìµœê³ ', 14))
    if len(commons) == 5:             # ë‹¨ì–´ ìˆ˜ê°€ ì¶©ë¶„í•˜ì—¬ 5ê°œê°€ ì˜ ë‚˜ì˜¨ ê²½ìš°ë§Œ ì‚¬ìš©
      for k in range(5):
        b.append(commons[k][0])
      words.append(b)
    else:                            
      df = df.drop(i)
  df['keys'] = words                  # ë‹¨ì–´ë“¤ë§Œ ëª¨ì•„ì„œ Keys ì¹¼ëŸ¼ì— ì €ì¥
  return df
# update_tokens í•¨ìˆ˜ë¥¼ ë°›ì•„ì„œ ì…ë ¥í•œ í‚¤ì›Œë“œì™€ ì§€ì—­ì´ ì¼ì¹˜í•˜ëŠ” ìŒì‹ì ì„ ì°¾ì•„ë‚´ëŠ” í•¨ìˆ˜, ë§ˆì°¬ê°€ì§€ë¡œ ìµœì¢… í•¨ìˆ˜ì—ì„œ ë°›ì•„ì„œ ì‚¬ìš©í•  ê²ƒì„
def matched_store():
  df = update_tokens()
  while True:
    # í‚¤ì›Œë“œ 3ê°œ ì…ë ¥ë°›ê¸°
    keywords=[]
    for i in range(3):
      keywords.append(input('ì›í•˜ëŠ” í‚¤ì›Œë“œ {}: '.format(i+1)))
    
    # í‚¤ì›Œë“œì™€ ë§ëŠ” ë°ì´í„° ì°¾ì•„ì„œ ë¦¬ìŠ¤íŠ¸ë¡œ ì €ì¥ 
    match = []
    for i in df.index:
      # í‚¤ì›Œë“œ 1,2,3 ì´ keys ì—´ì˜ ë‹¤ì„¯ ë‹¨ì–´ì— ëª¨ë‘ í¬í•¨ëœë‹¤ë©´ ê·¸ ì‹ë‹¹ ì´ë¦„ì„ ì°¾ì•„ matchì— ì €ì¥
      if (keywords[0] in df['keys'][i]) & (keywords[1] in df['keys'][i]) & (keywords[2] in df['keys'][i]):
        match.append(df['ìƒí˜¸ëª…'][i])     
        return df, match, keywords 

    else:                                
      print('í•´ë‹¹ ì¡°ê±´ì˜ ì‹ë‹¹ì€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.')
    
    a = input('ìŒì‹ì ì„ ë‹¤ì‹œ ê²€ìƒ‰í•˜ì‹œê² ìŠµë‹ˆê¹Œ? ë„¤ or ì•„ë‹ˆì˜¤: \n')
    # í”„ë¡œê·¸ë¨ ì¢…ë£Œ ì¡°ê±´
    if a == 'ì•„ë‹ˆì˜¤':
      print('í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.')     
      break     
# íŒŒì´ì°¨íŠ¸ ê·¸ë¦¬ê¸° ìœ„í•œ ê°’ë“¤ ê³„ì‚°í•˜ê¸°
def for_pie_chart():
  df, match, keywords = matched_store()         
  d = {}
  for name in match:                             
    a = df[df['ìƒí˜¸ëª…'] == name]
    a.index = range(0,1)
    label = keywords + ['ê¸°íƒ€']                  # label = í‚¤ì›Œë“œ ì„¸ê°œì™€ ë‚˜ë¨¸ì§€ ë‹¨ì–´
    ratio = []
    tok = a.loc[0, 'tokens']                     # tok = 'tokens' ì—´ì— ì €ì¥ëœ í† í° ë¦¬ìŠ¤íŠ¸ ê°’ë§Œ ê°€ì ¸ì˜¤ê¸° 
    left = len(tok)                             
    for key in keywords:
      num = tok.count(key)
      ratio.append(num)                          # í‚¤ì›Œë“œ ê°œìˆ˜ ì„¸ê¸°
      left -= num
    ratio.append(left)                           # ê¸°íƒ€ ë‹¨ì–´ë“¤ì˜ ê°œìˆ˜ ì„¸ê¸°
    d[name] = [ratio, label]
  return d                                       # ì‹ë‹¹ ì´ë¦„ì„ ë”•ì…”ë„ˆë¦¬ í‚¤ë¡œ, label ê³¼ ratio ëŠ” ê°’ìœ¼ë¡œ ì €ì¥í•´ì„œ return             

# ìµœì¢…í•¨ìˆ˜
def draw_pie_chart():
  d = for_pie_chart()
  for key in d:   # ì¡°ê±´ì— í•´ë‹¹í•˜ëŠ” ìŒì‹ì ì„ ëª¨ë‘ ì‹œê°í™”
    plt.rc('font', family='NanumBarunGothic') 
    explode = [0.05, 0.05, 0.05, 0.05] 
    colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
    plt.pie(d[key][0], labels=d[key][1], autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)  
    plt.title(key, size = 15)
    plt.show()

def program_5():
  while 1:
    mode_5 = int(input('ëª¨ë“œë¥¼ ê³¨ë¼ì£¼ì„¸ìš”: 1 - ë¦¬ë·° í…ìŠ¤íŠ¸ íŒŒì´ì°¨íŠ¸ ëª¨ë“œ, 2 - í”„ë¡œê·¸ë¨ ì¢…ë£Œ \n'))
    if mode_5 == 1:          
      draw_pie_chart()
    elif mode_5 == 2:
      break
```
